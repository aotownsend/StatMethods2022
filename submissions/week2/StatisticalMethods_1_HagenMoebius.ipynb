Now I make the simulation for part 1.4

```python
def coin():  #coin toss with fair coin
    b=np.random.rand(1)
    if b<0.5:
        #print("tails")
        return 1
    else:
        #print("heads")
        return 0
```

```python
def u_coin():  #coin toss with unfair coin
    b=np.random.rand(1)
    if b<1:
        #print("tails")
        return 1
    else:
        #print("heads")
        return 0
```

```python
def choose_coin(Iterations):    #choose coin and toss chosen coin
    
    fair_coin_number_of_tails=0
    unfair_coin_number_of_tails=0
    
    for i in range(Iterations):
           
        b=np.random.rand(1)
        if b<0.9:
            a=coin()
            if a==1:
                fair_coin_number_of_tails=fair_coin_number_of_tails+1
        else:
            a=u_coin()
            if a==1:
                unfair_coin_number_of_tails=unfair_coin_number_of_tails+1
    
    return (Iterations,fair_coin_number_of_tails,unfair_coin_number_of_tails)
        
    
        
        
```

```python
choose_coin(200000)   #200000 coin tosses
```

Just for my interest I plot the simulation result for different number of coin tosses

```python
def evolution(lower_limit,upper_limit,stepsize):
    E=np.zeros(0)
    step=np.zeros(0)
    for i in np.arange(lower_limit,upper_limit,stepsize):
        A=choose_coin(i)
        probability=A[2]/(A[1]+A[2])
        E=np.append([E],[probability])
        step=np.append([step],[i])
    
    return step,E     #step= number of tosses, E= probability that the coin was unfair when I got tails
        
```

```python
Test=evolution(10,100000,100)
```

```python
fig, ax = plt.subplots()
fig.set_figheight(9)
fig.set_figwidth(16)
#plt.scatter(Test[0],Test[1],s=1)
plt.plot(Test[0],Test[1])
plt.xlabel("Number of coin tosses that were simulated ")
plt.ylabel("probability that the coin was unfair when I got tails")
plt.hlines(0.1818,0,100000,"r",label="Analytical result of p=0.1818")
plt.legend(loc="upper right")
```
