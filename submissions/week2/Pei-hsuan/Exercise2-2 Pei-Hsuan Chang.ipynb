# Read and study data

```python
import math as math
import numpy as np
import csv
import matplotlib.pyplot as plt

file = open('us_marriages_divorces.csv')
reader = csv.reader(file)
data_list = list(reader)
file.close()
# read the csv file and put the data in one list
# print(data_list)

data_list.remove(data_list[0])
# remove the string part
```

### 1.

```python
y = [] # variable of year
m = [] # variable of marriages
d = [] # variable of divorces
p = [] # variable of population
mpp = [] # variable of marriages per 1000 population
dpp = [] # variable of divorces per 1000 population

for r in data_list: # and transform the data type
    y.append(int(r[0])) 
    m.append(float(r[1])) 
    d.append(float(r[2]))
    p.append(float(r[3]))
    mpp.append(float(r[4]))
    dpp.append(float(r[5]))
    
y_array = np.array(y)
m_array = np.array(m)
d_array = np.array(d)
p_array = np.array(p)
mpp_array = np.array(mpp)
dpp_array = np.array(dpp)

# print(y, m, d, p, mpp, dpp)
```

### 2.

```python
x = y_array 
y1 = m_array
y2 = d_array
# make the marriages and divorces as functions of years

plt.xlabel("Years")
plt.ylabel("Marriages(blue) / Divorces(green)")
plt.title("Marriages/Divorces v.s. Years")
plt.xlim(1860,2020)
plt.ylim(10000,2700000)
plt.plot(x, y1, c='blue', label = "Maariages")
plt.plot(x, y2, c='green', label = "Divorces")
plt.legend(loc="upper left")
plt.show()
```

### 3.

```python
i = np.nanargmax(mpp) 
j = np.nanargmax(dpp)
# find the index that the maximum happens in the array, ignore nan value

print(y[i], ': the year has maximum of marriages per 1000 people')
print(y[j], ': the year has maximum of divorces per 1000 people')
```

### 4.

```python
for i in range(len(mpp)):
    index = np.argmax(mpp)    
    if mpp[index] <= 20:
        break
    if i == index:
        mpp[i] = 0
        
while 0 in mpp:
    mpp.remove(0)
# remove nan from the list

mpp_new = np.array(mpp)

plt.title("Marriages per 1000 people", y =-0.15)
plt.ylabel("Frenquency")
plt.hist(mpp_new, bins = 10, color ='c', edgecolor="k")

plt.show()

mean = np.mean(mpp_new)
median = np.median(mpp_new)
std = np.std(mpp_new)
mad = np.median(abs(mpp_new-np.median(mpp_new)))
nmad = mad*1.483
# calculate mean, median, std, nmad

print(" median = ",median,"\n","mean = ",mean,"\n","std = ", std,"\n", "nmad = ", nmad)
```
