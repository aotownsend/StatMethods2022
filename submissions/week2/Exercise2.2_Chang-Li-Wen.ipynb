```python
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
file=open('us_marriages_divorces.csv')

data = list(csv.reader(file)) # read this file and put the elements in the list called data
file.close()
data.remove(data[0])
```

```python
y=[]
m=[]
d=[]
p=[]
mp1000=[]
dp1000=[]
for i in data:
    y.append(int(i[0]))
    m.append(float(i[1]))
    d.append(float(i[2]))
    p.append(float(i[3]))
    mp1000.append(float(i[4]))
    dp1000.append(float(i[5]))
```

```python
y_np=np.array(y)
m_np=np.array(m)
d_np=np.array(d)
p_np=np.array(p)
mp1000_np=np.array(mp1000)
dp1000_np=np.array(dp1000)
```

```python
plt.plot(y_np,m_np, label='marriages')
plt.plot(y_np,d_np, label='divorces')
plt.xlabel("year")
plt.ylabel("number")
plt.legend(loc='upper left')
```

```python
print("maximum of marriages per 1000 people occurs at " + str(y[np.nanargmax(mp1000)]))
print("maximum of divorces per 1000 people occurs at " + str(y[np.nanargmax(dp1000)]))
```

```python
mp1000[115]=999
mp1000[116]=999
mp1000[128]=999
mp1000[129]=999
while 999 in mp1000:
    mp1000.remove(999)
mp1000_np=np.array(mp1000)
#print(mp1000_np)
```

```python
mp1000_np.sort()
#print(mp1000)

median=mp1000_np[70]
print("median is: "+str(median))


mean=np.mean(mp1000_np)
print("mean is: "+str(mean))

std=np.std(mp1000_np)
print("standard deviation is: "+str(std))

mp1000_npmm = mp1000_np-median #mp1000_np minus median
#print(mp1000_npmm)
i=0
while i<141:
    mp1000_npmm[i]=abs(mp1000_npmm[i])
    i+=1
#print(mp1000_npmm)

mp1000_npmm.sort()
#print(mp1000_npmm)
MAD=mp1000_npmm[70]
#print("MAD is: "+ str(MAD))
nmad=MAD*1.483
print("nmad is: "+str(nmad))
```

```python
p30s=[]
p2000s=[]
i=0
while i<10 :
    p30s.append(p[63+i])
    p2000s.append(p[133+i])
    i+=1
#print(p30s)
#print(p2000s)
print("the average US population during the 30s is: " + str(np.mean(p30s)))
print("the average US population during the 2000s is: " + str(np.mean(p2000s)))
```

```python
plt.hist(mp1000_np,bins=10, alpha=0.6)
plt.axvline(mean,color='tab:blue',linestyle='dashed', label='mean')
plt.axvline(mean+std,color='tab:blue',linestyle='dotted', label='std(around mean)')
plt.axvline(mean-std,color='tab:blue',linestyle='dotted')
plt.axvline(median,color='tab:red',linestyle='solid', label='median')
plt.axvline(median+nmad,color='tab:red',linestyle='dashdot', label='nmad(around median)')
plt.axvline(median-nmad,color='tab:red',linestyle='dashdot')
plt.legend()
plt.xlabel("Marriages per year per 1000 people in the US between 1867 and 2011")
plt.ylabel("Frequency")
plt.figtext(0.5, -0.04, "Figure 1: plot to reproduce", ha='center', fontsize=12) #(, , caption, , )
```
