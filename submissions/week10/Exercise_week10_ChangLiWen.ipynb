```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import math
import random
from scipy.stats import gaussian_kde
from matplotlib.ticker import MaxNLocator
from numpy import random
```

```python
def poisson(l,n):
    x=(l**n)*math.exp(-l)/math.factorial(n)
    return x
```

```python
print(poisson(1,2))
```

```python
def upper_limit(l):
    i=0
    sum=0
    while i<100:
        sum=sum+poisson(l,i)
        if sum>0.9:
            y=i
            break
        i=i+1
    return y
```

```python
above=[]
x=np.linspace(1,12,12)
for i in x:
    above.append(upper_limit(i))
plt.plot(above,x,'o--')
print(above)
```

```python
def lower_limit(l):
    i=0
    sum=0
    while i<100:
        sum=sum+poisson(l,i)
        if sum>0.1:
            y=i
            break
        i=i+1
    return y
```

```python
below=[]
for i in x:
    below.append(lower_limit(i))
plt.plot(below,x,'o--')
print(below)
```

```python
ax = plt.figure().gca()
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
plt.xlim([-1,18])
plt.plot(above,x,'o--')
plt.plot(below,x,'o--')
```

```python
def middle_left(l):
    i=0
    sum=0
    while i<100:
        sum=sum+poisson(l,i)
        if sum>0.05:
            y=i
            break
        i=i+1
    return y
```

```python
def middle_right(l):
    i=0
    sum=0
    while i<100:
        sum=sum+poisson(l,i)
        if sum>0.95:
            y=i
            break
        i=i+1
    return y
```

```python
between_left=[]
between_right=[]
for i in x:
    between_left.append(middle_left(i))
    between_right.append(middle_right(i))
print(between_left)
print(between_right)
```

```python
ax = plt.figure().gca()
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
plt.xlim([-1,19])
plt.plot(above,x,'ro--',label="10% above")
plt.plot(below,x,'bo--',label="10% below")
plt.plot(between_left,x,'go--',label="between=90%")
plt.plot(between_right,x,'go--')

plt.xlabel("number of observed events")
plt.ylabel("expectation value")
plt.axvline(x = 5, color = 'orange',label="n=5")
plt.legend()
```

# B-2

```python
def fivepercent(l):
    sum=0
    i=0
    while i<6:
        sum=sum+poisson(l,i)
        i=i+1
    return sum
```

```python
x=np.linspace(10,11,100001)
for i in x:
    if fivepercent(i)<0.05:
        print(i-1.8)
        break
print(fivepercent(10.51304))
```
