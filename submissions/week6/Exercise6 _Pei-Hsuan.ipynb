```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample1 = []

with open('test.txt', 'r') as dat_file:
    for row in dat_file:
        row_r = float(row)
        sample1.append(row_r)

dat_file.close()
sample = sample1[:10000]
n = len(sample)
mu = sum(sample)/n

likehood = np.zeros(n)
plt.figure(figsize=(8, 6)) # show how the value distribute
plt.hist(sample,bins=100)
plt.show()
```

```python
def pro(x, alpha):
    return (x)**(-alpha)

likelihood = np.zeros(100)
x = np.linspace(0.75, 2.55, num = len(sample) )
a = np.linspace(1, 6, num = 100)
    

for indx, value in enumerate(a):
    
    prob = 0
    for j in x:
        prob += np.log((value)*pro(j, value))
    prob = prob-(value)
#     print(prob, indx)
    
    likelihood[indx] = prob
```

```python
max_ = max(likelihood)
min_ = min(likelihood)
print(max_, min_)
max1 = max(sample)
likelihood2 = [abs((x-min_)/(max_-min_)) for x in likelihood]
```
