```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
file = open('exercise6.csv')
read=csv.reader(file)
data = list(read)
file.close()
d=[]
for i in data:
    d.append(float(i[0]))
plt.hist(d, bins=80)
```

```python
dd=np.array(d)
max=np.nanargmax(dd)
print(dd[max])
```

```python
ls=0 #ls=log sum
for i in dd:
    ls=ls+np.log(i)
print(ls)
```

```python
p=np.linspace(11.4,11.48,100)
```

```python
ll=[]
for i in p:
    ll.append(i*ls-(dd[max]**(i+1))/(i+1))
plt.plot(p,ll)
plt.ylabel("lnL")

print(p[np.nanargmax(ll)])
```

```python
twoll=[]
for i in ll:
    if i>=ll[np.nanargmax(ll)]-0.5:
        twoll.append(i*0.999)
    else:
        twoll.append(i)
plt.plot(p,twoll)
```

```python
leftll=[]
rightll=[]
i=0
while i<100:
    if i<np.nanargmax(ll):
        leftll.append(twoll[i])
    else:
        rightll.append(twoll[i])
    i+=1
print(np.nanargmax(ll))
print(p[np.nanargmax(ll)])
    
```

```python
plt.plot(np.linspace(11.4,p[46],47), leftll)
plt.plot(np.linspace(p[47],11.48,53), rightll)
```

```python
print(p[np.nanargmax(leftll)])
print(p[np.nanargmax(rightll)])
print(np.nanargmax(leftll))
print(np.nanargmax(rightll))
```

```python
plt.plot(p,ll)
plt.axvline(x = p[np.nanargmax(leftll)+1], color = 'orange', linestyle='dashed', label = 'std')
plt.axvline(x = p[np.nanargmax(rightll)+46], linestyle='dashed', color = 'orange')
plt.axhline(y = ll[np.nanargmax(ll)]-0.5, linestyle='dashed', color = 'red', label = 'lnL-0.5')
plt.legend()
plt.ylabel("lnL")
print(p[np.nanargmax(ll)])
print(p[np.nanargmax(leftll)+1])
print(p[np.nanargmax(rightll)+46])
```
