I define a function to make the binning easier

```python


def cutting(start_time,end_time,Year):
    cut=np.zeros(len(Year))
    for i in range(len(Year)):
        if Year[i]>=start_time and Year[i]<=end_time:
            cut[i]=1
        
    return cut
```

Now we want to calculate the weighted mean

The weighted mean is defined by: $$\frac{\sum_{i=1}^{n} w_i \cdot x_i}{\sum_{i=1}^{n} w_i }$$

In general gaussian error propagation tells us: $$u=\sqrt{\Big(\frac{df}{dx_1}\Big)^2\cdot u_{x_1}^2+\Big(\frac{df}{dx_2}\Big)^2\cdot u_{x_2}^2+...}$$

This implies that the uncertainty of the weighted mean will be:
$$\frac{1}{{\sum_{i=1}^{n} w_i }}\cdot \sqrt{{\sum_{i=1}^{n} w_i^4 }}$$

```python
def weighted_mean(cut,value,uncertainty,year,month):
    cutted_value=cut*value
    cutted_uncertainty=cut*uncertainty
    cutted_year=cut*year
    cutted_month=cut*month
    
    cut_time=cutted_year+1/12*cutted_month
    time_mean=np.mean((cut_time))*(len(cut)/np.sum(cut))
    
    weighted_mean=(np.nansum(cutted_value*cutted_uncertainty))/(np.nansum(cutted_uncertainty))
    
    weighted_uncertainty=1/np.nansum(cutted_uncertainty)*np.sqrt(np.nansum(cutted_uncertainty**2))
    
    return weighted_mean, time_mean, weighted_uncertainty
```
