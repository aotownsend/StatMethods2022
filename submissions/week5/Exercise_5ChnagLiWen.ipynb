```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random
```

```python
n=10
i=0
a=[]
b=[]
while i<1000: #repeat the exp 1000 times
    x=[]
    y=[]
    yerr=[]
    yerr = random.normal(loc=0, scale=0.05, size = n) #create error-list
    j=0
    while j<n: #create x and y list
        x.append(random.random())
        y.append(1+0.5*x[j]+yerr[j])
        j+=1
    
    product=0 
    k=0
    while k<n:
        product = product + x[k]*y[k]
        k+=1
    s = (product/n-np.mean(x)*np.mean(y))/(np.var(x)) #slope
    intercept = np.mean(y)-s*np.mean(x) #intercept
    a.append(s)
    b.append(intercept)
    
    i+=1
#fig, (ax1, ax2) = plt.subplots(2)
#ax1.hist(a, bins=15, label="a")
#ax2.hist(b, bins=15, label="b")
plt.hist(a, bins=13, label='a (slope)')
plt.hist(b, bins=13, label='b (intercept)')
plt.legend(loc='upper left')
plt.xlabel("Value")
plt.ylabel("Frequency")
print("a is ", np.mean(a), "and b is ", np.mean(b))
```

```python
# we have x-error now
n=100
i=0
a=[]
b=[]
while i<1000: #repeat the exp 1000 times
    x=[]
    xr=[]
    y=[]
    yerr=[]
    yerr = random.normal(loc=0, scale=0.2, size = n) #create y-error
    xerr = random.normal(loc=0, scale=0.2, size = n) #create x-error
    j=0
    while j<n: #create x and y list
        xr.append(random.random())
        y.append(1+0.5*xr[j]+yerr[j])
        x.append(xr[j]+xerr[j])
        j+=1
    
    product=0 
    k=0
    while k<n:
        product = product + x[k]*y[k]
        k+=1
    s = (product/n-np.mean(x)*np.mean(y))/(np.var(x)) #slope
    intercept = np.mean(y)-s*np.mean(x) #intercept
    a.append(s)
    b.append(intercept)
    
    i+=1
#fig, (ax1, ax2) = plt.subplots(2)
#ax1.hist(a, bins=15, label="a")
#ax2.hist(b, bins=15, label="b")
plt.hist(a, bins=13, label='a (slope)')
plt.hist(b, bins=13, label='b (intercept)')
plt.legend(loc='upper left')
plt.xlabel("Value")
plt.ylabel("Frequency")
print("a is ", np.mean(a), "and b is ", np.mean(b))
```
