```python
import numpy as np
import pandas as pd
import math as m

import matplotlib.pyplot as plt
from scipy.special import gamma
np.random.seed(523)

def f1(x):
  p = 1/np.sqrt(2*m.pi)*np.exp(-x*x/2) 
  return  p

mode = 0
c = f1(0)

def gaus_gen(n):
  i = 0
  output = np.zeros(n)
  while i < n:
    U = np.random.uniform(size = 1)
    V = np.random.uniform(-3,3,size = 1)
    if U < 1/c * f1(V):
      output[i] = V
      i = i + 1
  return output


px = np.arange(-3,3,0.01)
py = f1(px)


Y = gaus_gen(n = 100)
fig,ax = plt.subplots()
temp = ax.hist(Y,bins='auto',density=True)
ax.plot(px,py)
plt.title("Gaus")
plt.show()
```
